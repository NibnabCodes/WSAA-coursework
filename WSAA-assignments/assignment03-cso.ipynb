This program retrieves the dataset for the "exchequer account (historical series)" taken from the [CSO](https://data.cso.ie/#) and stores it into a file called "cso.json"

```python
# imports
import requests
import os
```

```python
urlStart = "https://ws.cso.ie/public/api.restful/PxStat.Data.Cube_API.ReadDataset/"
urlEnd = "/JSON-stat/2.0/en"  
```

```python
def getAllAsFile(dataset):
    url = urlStart + dataset + urlEnd
    response = requests.get(url)
    filepath = os.path.join(os.path.dirname(os.path.abspath("__file__")), "cso.json")
    with open(filepath, "w", encoding="utf-8", newline='\n') as f:
        f.write(response.text)
```

```python
if __name__ == "__main__":
    getAllAsFile("FIQ02")
```
