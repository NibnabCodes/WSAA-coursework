This program retrieves the dataset for the "exchequer account (historical series)" taken from the [CSO](https://data.cso.ie/#) and stores it into a file called "cso.json"

```python
# imports
import requests
import os
```

```python
# Base URL components for the CSO PxStat API
urlStart = "https://ws.cso.ie/public/api.restful/PxStat.Data.Cube_API.ReadDataset/"
urlEnd = "/JSON-stat/2.0/en"  
```

```python
# Retrieve dataset from CSO API and save to cso.json
def getAllAsFile(dataset):
    url = urlStart + dataset + urlEnd
    response = requests.get(url)
    # Build file path relative to the current directory
    # Reference: https://docs.python.org/3/library/os.path.html
    filepath = os.path.join(os.path.dirname(os.path.abspath("__file__")), "cso.json")
    # newline='\n' prevents Windows adding ^M carriage return characters
    # Reference: https://www.pythonmorsels.com/newlines-and-escape-sequences/
    with open(filepath, "w", encoding="utf-8", newline='\n') as f:
        f.write(response.text)
```

```python
# Run program only when executed directly, not when imported as a module
# Reference: https://www.geeksforgeeks.org/python/what-does-the-if-__name__-__main__-do/
if __name__ == "__main__":
    getAllAsFile("FIQ02")
```
