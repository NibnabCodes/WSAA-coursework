**Assignment04-Github**

This program interacts with the GitHub API to automate editing a file in a repository.
It reads a `.txt` file from the WSAA-coursework repository, replaces all instances of the name "Andrew" with "Niamh", and pushes the updated file back to the repository.
GitHub authorisation is handled via a Personal Access Token stored securely in a `config.py` file, which is excluded from version control via `.gitignore`.

```python
import github # GitHub API library (PyGithub)
from github import Github, Auth # imports Github class and Auth module from PyGithub
import sys # standard library for system/runtime info

sys.path.append("..") # Allows import of config file from parent directory
from config import config as cfg # Loads GitHub API key securely 
```

```python
# Authenticate with GitHub using fine grained Personal Access Token
# Reference: https://docs.github.com/en/authentication/keeping-your-account-and-data-secure/managing-your-personal-access-tokens
g = Github(auth=Auth.Token(cfg["githubkey"]))
```

```python
# Access the repository by name and print its clone URL
repo = g.get_repo("NibnabCodes/WSAA-coursework") 
print(repo.clone_url) 
```

```python
# Relative path to the target file within the repo
file_path = "WSAA-assignments/assignment4.txt"

# GitHub branch to read from and commit changes to
branch = "main"
```

```python
# Fetch file from GitHub and decode content from bytes to string
# Reference: https://pygithub.readthedocs.io/en/latest/github_objects/ContentFile.html
file = repo.get_contents(file_path, ref=branch) # Fetches file object from the repo and branch
file_content = file.decoded_content.decode("utf-8") # Converts file content from bytes to string using UTF-8 encoding
```

```python
# Replace 'Andrew' with 'Niamh' in the file content
# Reference: https://www.w3schools.com/python/ref_string_replace.asp
updated_content = file_content.replace("Andrew", "Niamh")
```

```python
# Commit and push updated file back to the repository
# sha identifies the current version of the file to avoid conflicts
# Reference: https://pygithub.readthedocs.io/en/latest/github_objects/Repository.html
repo.update_file(
    path=file_path,
    message="Replace Andrew with Niamh",
    content=updated_content,
    sha=file.sha,
    branch=branch
)
```
